```python
import csv
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
dataframe = Path(r"C:\Users\DELL\Desktop\assignment\new_housing_prices.csv")
```

```python
housing_prices_affordability_df= pd.read_csv(dataframe)
housing_prices_affordability_df
```

```python
housing_prices_affordability_df1= housing_prices_affordability_df.dropna()
housing_prices_affordability_df1
```

```python
housing_prices_affordability_df1.set_index('date')
```

```python
x = housing_prices_affordability_df1["date"]
y1 = housing_prices_affordability_df1["New housing price index"]
y2 = housing_prices_affordability_df1["affordability index"]
```

```python
# using subplots() function
fig, ax = plt.subplots(figsize = (20, 17))
plt.title('Comparison between Housing prices & affordability index')

# set xticks rotation before creating ax2
plt.xticks(rotation=90)

# using the twinx() for creating another
# axes object for secondary y-Axis
ax2 = ax.twinx()
ax.bar(x, y1, color = 'yellow')
ax2.plot(x, y2, color = 'red')

# giving labels to the axises
ax.set_xlabel('date', color = 'black')
ax.set_ylabel('New Housing price index', color = 'black')

# secondary y-axis label
ax2.set_ylabel('affordability index', color = 'red')

plt.tight_layout()

plt.show()
```
