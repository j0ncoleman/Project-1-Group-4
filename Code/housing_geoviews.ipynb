```python
combined_Geoview_df = pd.concat(
    [coordinates_df, Rental_df], 
    join = "inner",
    sort = True
)
```

```python
combined_Geoview_df.head()
```

```python
lat = coordinates_df['Latitude']
lon = coordinates_df['Longitude']
Housing_price = Rental_df['Price Januray 2022']
```

```python
combined_Geoview_df['lat'] = lat
combined_Geoview_df['lon'] = lon
combined_Geoview_df['Housing_price'] = Housing_price
```

```python
combined_Geoview_df
```

```python
combined_Geoview = combined_Geoview_df.head(10)
```

```python
combined_Geoview
```

```python
map = folium.Map(location=[56.415211, -98.739075], zoom_start=6)

for row in combined_Geoview.itertuples():
    folium.Marker([row.lat, row.lon], popup=str(row.Housing_price)).add_to(map)

map.save("housing_price_map.html")
```
